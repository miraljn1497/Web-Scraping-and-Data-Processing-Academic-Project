# Web Scraping BooksToScrape

In the notebook, we scrape book data from https://books.toscrape.com.
The goal is to collect structured data that will later be stored in a database,
visualized, used for machine learning modeling.

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

Before scraping, we verified that books.toscrape.com allows scraping.
This website is specifically designed for educational purposes.

```python
url = "https://books.toscrape.com/catalogue/page-1.html"
response = requests.get(url)
soup = BeautifulSoup(response.text, "html.parser")

books = soup.find_all("article", class_="product_pod")
len(books)
```

```python
data = []

for book in books:
    title = book.h3.a["title"]
    price = book.find("p", class_="price_color").text.replace("£", "")
    availability = book.find("p", class_="instock availability").text.strip()
    rating = book.p["class"][1]  # e.g. Three, Four
    
    data.append([title, price, availability, rating])
```

```python



df = pd.DataFrame(data, columns=["title", "price", "availability", "rating"])
df.head()
```

```python
all_books = []

for page in range(1, 51):
    url = f"https://books.toscrape.com/catalogue/page-{page}.html"
    soup = BeautifulSoup(requests.get(url).text, "html.parser")
    books = soup.find_all("article", class_="product_pod")
    
    for book in books:
        title = book.h3.a["title"]
        price = book.find("p", class_="price_color").text.replace("£", "")
        availability = book.find("p", class_="instock availability").text.strip()
        rating = book.p["class"][1]
        all_books.append([title, price, availability, rating])
```

```python
import os

os.makedirs("data", exist_ok=True)

df_books = pd.DataFrame(
    all_books, 
    columns=["title", "price", "availability", "rating"]
)

df_books.to_csv("data/books_raw.csv", index=False)
df_books.shape
```
