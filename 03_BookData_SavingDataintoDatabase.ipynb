The objective of this notebook is to load the scraped book data from CSV files
and store it into a SQLite database. This allows efficient data storage,
querying, and integration with visualization and machine learning tools.

SQLite is used because it is lightweight, easy to use, and does not require
a separate database server

```python
import pandas as pd
import sqlite3
import os
```

## Load CSV Data
The scraped book data is loaded from the CSV file created in the web scraping step.

```python
csv_path = "data/books_raw.csv"
df_books = pd.read_csv(csv_path)

df_books.head()
```

## Create SQLite Database
We create a SQLite database file inside the `data` folder and establish a connection to it.


```python
db_path = "data/books_database.db"
conn = sqlite3.connect(db_path)
```

## Save Data into Database Table
The Pandas DataFrame is stored as a table in the SQLite database.
If the table already exists, it will be replaced.

```python
df_books.to_sql(
    name="books",
    con=conn,
    if_exists="replace",
    index=False
)
```

## Verify Stored Data
We query the database to ensure that the data has been saved correctly.

```python
query = "SELECT * FROM books LIMIT 5;"
df_check = pd.read_sql(query, conn)
df_check
```

## Database Table Information
This step helps us understand the structure of the database table.

```python
table_info = pd.read_sql("PRAGMA table_info(books);", conn)
table_info
```

## Close Database Connection
The database connection is closed after the operations are completed.


```python
conn.close()
```

## Summary
In this notebook, the scraped book data was successfully loaded from a CSV file
and stored into a SQLite database. The database contains a table named `books`
with all relevant book information.
